```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, scale
from os import listdir
import keras
import squarify
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop
from sklearn.model_selection import train_test_split
pd.set_option('display.max_rows', 1000) 
```

```python
pd.options.display.max_columns = None
frames = []
def find_csv_filenames( path_to_dir, suffix=".csv" ):
    filenames = listdir(path_to_dir)
    return [ filename for filename in filenames if filename.endswith( suffix ) ]
filenames = find_csv_filenames("premier_league")
for name in filenames:
    try:
        name = 'premier_league/' + name
        df = pd.read_csv(name, engine='python')
        #print('The shape is' + str(df.shape) + 'in the file' + name)
        df = (df[['Div', 'HomeTeam', 'AwayTeam',
                'FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG',
                'HTR', 'HS', 'AS', 'HST', 'AST',
                'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR']])
        #print('The new shape is' + str(df.shape) + 'in the file' + name)
        frames.append(df)
    except Exception as error:

        print(error)
df1 = pd.concat(frames)
#a = pd.read_csv("premier_league/2017_2018.csv")
#b = pd.read_csv("premier_league/2016_2017.csv")
#frames = [a, b]
#df1 = pd.concat(frames)
#print(a.shape)
#print(b.shape)
df1.head()
```

```python
#Home teams played array
ht = []
#Home team wins array
htw = []

df1['home_team_chance_to_win_home'] = ''
df1['home_team_chance_to_win_agaisnt_away'] = ''
for home_team, away_team in zip(df1['HomeTeam'].unique(), df1['AwayTeam'].unique()):
    # Times each team played as home team
    played_ht = len(df1.loc[df1['HomeTeam'] == home_team, ['HomeTeam', 'FTR']])
    ht.append(played_ht)
    # Times each team won as a home team
    played_htw = len(df1.loc[(df1['HomeTeam'] == home_team) & (df1['FTR'] == 'H'), ['HomeTeam', 'FTR']])
    if str(home_team).lower() == 'nan':
        continue
    #print('Percentage of', home_team, 'winning is' , str( "%.2f" % ((played_htw / played_ht) * 100) + '%'), 'playing at home')
    htw.append(played_htw)
    home_team_win_percentage = df1.loc[df1['HomeTeam'] == home_team, ['home_team_chance_to_win_home']] = float("%.2f" % ((played_htw / played_ht) * 100))
    home_team_played_agaisnt_away = df1.loc[(df1['HomeTeam'] == home_team) & (df1['AwayTeam'] == away_team), ['HomeTeam', 'AwayTeam', 'FTR']]
    #print(home_team_played_agaisnt_away)
    home_team_win_agaisnt_away = df1.loc[(df1['HomeTeam'] == home_team) & (df1['AwayTeam'] == away_team) & (df1['FTR'] == 'H'), ['HomeTeam', 'AwayTeam', 'FTR']]
    #print(home_team_win_agaisnt_away)
    if len(home_team_played_agaisnt_away) == 0:
        continue
    home_team_win_percentage = df1.loc[(df1['HomeTeam'] == home_team) & (df1['AwayTeam'] == away_team) & (df1['FTR'] == 'H'), ['home_team_chance_to_win_agaisnt_away']] = len(home_team_win_agaisnt_away) / len(home_team_played_agaisnt_away) * 100
    
    #print(home_team_win_percentage)
    #home_team_win_percentage = df1.loc[(df1['HomeTeam'] == home_team) & (df1['AwayTeam'] == away_team) & (df1['FTR'] != 'H'), ['home_team_chance_to_win_agaisnt_away']] = 0


nulify_percentage = df1.loc[df1['home_team_chance_to_win_agaisnt_away'] == '', ['home_team_chance_to_win_agaisnt_away']] = 0
#df1.head(20)
```

```python
games = df1
games['HTR'].replace('H', 2).replace('D', 1).replace('A', 0)
games['FTR'].replace('H', 2).replace('D', 1).replace('A', 0)
i = 0
for home_team in games['HomeTeam'].unique():
    games.loc[(games['AwayTeam'] == home_team), ['AwayTeam']] = i
    games.loc[(games['HomeTeam'] == home_team), ['HomeTeam']] = i
    i += 1
games.head()


#results = df1['HTR']
#games.corrwith(results)
```

```python
team_home = pd.get_dummies(df1['HomeTeam'], prefix='Home')
team_away = pd.get_dummies(df1['AwayTeam'], prefix='Away')
teams = pd.concat([team_home, team_away], axis=1)

df1 = pd.concat([df1, team_away], axis=1)
df1 = df1.drop(['HomeTeam', 'AwayTeam'], axis=1)
#df1.head()
```

```python
ftr = {'H': 0, 'D': 1, 'A': 2}
df1['FTR'] = df1['FTR'].map(ftr)
df1['HTR'] = df1['HTR'].map(ftr)
```

```python
df1.fillna(9999, inplace=True)
df1.dropna()

#df1.notnull().sum()
```

```python

win_home = np.sum(df1['FTR'] == 0)
draw = np.sum(df1['FTR'] == 1)
win_away = np.sum(df1['FTR'] == 2)
results_pie = [win_home, draw, win_away]
labels = ['Home', 'Draw', 'Away']
plt.pie(results_pie, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
df1 = df1.replace('E0', 0)
```

```python
X = df1.drop(['FTHG', 'FTAG'], axis=1)

Y = df1[['FTHG', 'FTAG']]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=42)
X_train = scale(X_train)
y_train = scale(y_train)
X_test = scale(X_test)
y_test = scale(y_test)
print(X_train.shape)
print(y_train.shape)
```

```python
model = Sequential()
```

```python
model.add(Dense(1500, activation='relu', input_shape=(60,)))
model.add(Dropout(0.5))
model.add(Dense(120, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['accuracy'])
```

```python

history = model.fit(X_train, y_train,batch_size=3420, epochs=150,verbose=2, validation_data=(X_test, y_test))
```

```python
scores = model.evaluate(X_test, y_test, verbose=2)
```

```python
print('Test Loss:', scores[0])
print('Test Accuracy', scores[1])
```
